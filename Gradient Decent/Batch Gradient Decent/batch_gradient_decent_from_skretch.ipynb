```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X,y = make_regression(n_features=1,n_informative=1,n_samples=4,n_targets=1,noise=80,random_state=4)
```

```python
X
```

```python
y
```

```python
print(X.shape)
print(y.shape)
```

```python
plt.scatter(X,y)
```

```python
reg = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#regression Line
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
#applying gradient decent for m = 79.47677385 and b = 0
y_pred = ((79.47677385*X.ravel())+0)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b = 0 GD')
plt.legend()
```

```python
#now applying gradient decent for some epochs
m = 79.47677385
b=0
loss_slope = -2*(np.sum(y-(m*X.ravel())-b))
```

```python
loss_slope
```

```python
learning_rate = 0.1
print(f"Step Size for b = {b}: {learning_rate*loss_slope}")
b = b - (learning_rate*loss_slope)
b
```

```python
y_pred1 = (79.47677385*X.ravel())+b 
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b = 0 GD')
plt.plot(X,y_pred1,color='blue',label=f'epoch-1 b = {b} GD')
plt.legend()
```

```python
#epoch 2
loss_slope = -2*np.sum(y-(m*X.ravel())-b)
print(f"Step Size for b = {b}: {learning_rate*loss_slope}")
b = b - (learning_rate*loss_slope)
y_pred2 = (79.47677385*X.ravel())+b 
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b = 0 GD')
plt.plot(X,y_pred1,color='blue',label=f'epoch-1 GD')
plt.plot(X,y_pred2,color='pink',label=f'epoch-2 b = {b} GD')
plt.legend()
```

```python
#epoch 3
loss_slope = -2*np.sum(y-(m*X.ravel())-b)
print(f"Step Size for b = {b}: {learning_rate*loss_slope}")
b = b - (learning_rate*loss_slope)
y_pred3 = (79.47677385*X.ravel())+b 
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b = 0 GD')
plt.plot(X,y_pred1,color='blue',label=f'epoch-1 GD')
plt.plot(X,y_pred2,color='pink',label=f'epoch-2 GD')
plt.plot(X,y_pred3,color='purple',label=f'epoch-3 b = {b} GD')
plt.legend()
```

```python
#now applying gradient decent for some epochs
m = 79.47677385
b=-100
learning_rate = 0.1
epochs = 10

for i in range(epochs):
    loss_slope = -2*(np.sum(y-(m*X.ravel())-b))
    b = b - (learning_rate*loss_slope)
    y_pred = (m * X) + b
    plt.plot(X,y_pred,label=f'epoch-{i}')
    plt.legend()
plt.scatter(X,y)
```
